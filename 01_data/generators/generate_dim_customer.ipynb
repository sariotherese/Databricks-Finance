# Generate dim_customer
Creates the customer dimension with Faker and writes data/dim_customer.csv.

```python
from pathlib import Path

import numpy as np
import pandas as pd
from faker import Faker

SEED = 42
COUNT = 1500
rng = np.random.default_rng(SEED)
np.random.seed(SEED)
Faker.seed(SEED)
fake = Faker()
fake.seed_instance(SEED)

def find_project_root() -> Path:
    current = Path.cwd().resolve()
    for candidate in [current, *current.parents]:
        if (candidate / 'README.md').exists() and (candidate / '01_data').exists():
            return candidate
    return current

first_names = [fake.first_name() for _ in range(COUNT)]
last_names = [fake.last_name() for _ in range(COUNT)]
cities = [fake.city() for _ in range(COUNT)]
emails = [f"{f.lower()}.{l.lower()}{i + 1}@examplebank.com" for i, (f, l) in enumerate(zip(first_names, last_names))]
join_dates = pd.to_datetime(rng.choice(pd.date_range('2018-01-01', '2026-06-30'), size=COUNT))

dim_customer = pd.DataFrame({
    'customer_key': np.arange(1, COUNT + 1, dtype=np.int32),
    'first_name': first_names,
    'last_name': last_names,
    'email': emails,
    'city': cities,
    'segment': rng.choice(['Mass', 'Affluent', 'Private'], size=COUNT, p=[0.65, 0.3, 0.05]),
    'join_date': join_dates,
})

output_dir = find_project_root() / '01_data' / 'raw'
output_dir.mkdir(parents=True, exist_ok=True)
output_path = output_dir / 'dim_customer.csv'
dim_customer.to_csv(output_path, index=False)

print(f'Rows: {len(dim_customer)}')
print(f'Wrote: {output_path}')
dim_customer.head()
```
