# Generate dim_transaction_type
Creates the transaction type dimension and writes data/dim_transaction_type.csv.

```python
from pathlib import Path

import numpy as np
import pandas as pd

SEED = 42
np.random.seed(SEED)

def find_project_root() -> Path:
    current = Path.cwd().resolve()
    for candidate in [current, *current.parents]:
        if (candidate / 'README.md').exists() and (candidate / '01_data').exists():
            return candidate
    return current

tx_types = [
    'Deposit',
    'Withdrawal',
    'Card Purchase',
    'Card Refund',
    'Transfer In',
    'Transfer Out',
    'Loan Payment',
    'Fee',
]

dim_transaction_type = pd.DataFrame({'transaction_type': tx_types})
dim_transaction_type.insert(0, 'transaction_type_key', np.arange(1, len(dim_transaction_type) + 1, dtype=np.int32))

output_dir = find_project_root() / '01_data' / 'raw'
output_dir.mkdir(parents=True, exist_ok=True)
output_path = output_dir / 'dim_transaction_type.csv'
dim_transaction_type.to_csv(output_path, index=False)

print(f'Rows: {len(dim_transaction_type)}')
print(f'Wrote: {output_path}')
dim_transaction_type.head()
```
